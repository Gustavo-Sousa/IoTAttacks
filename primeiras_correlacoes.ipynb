# Read first_dataset.csv generated by subdataset notebook 

```python
df = pd.read_csv('first_dataset.csv', low_memory=False, sep=',')
```

# Initial Analysis

## Calculate correlation between frame delta and packet lenght

```python
#There's exaclty one value in this column that goes beyond a  float64 which is weird but is ignored, probably it's the datatype
frame_delta = pd.to_numeric(df['frame.time_delta'], errors='coerce') 
packet_lenght = df['ipv6.plen']

correlation = frame_delta.corr(packet_lenght)
print(correlation)
```

### The correlation isn't here, so the packet lenght probably does not affect the delay between packets, which makes sense since the packet size should be more or less the same, if the message is very big it gets fragmented in several packets

## Calculate correlation between frame delta and message lenght

```python
frame_delta = pd.to_numeric(df['frame.time_delta'], errors='coerce') 
mqtt_lenght = pd.to_numeric(df['mqtt.len'], errors='coerce')

correlation = mqtt_lenght.corr(packet_lenght)
print(correlation)
```

### Now we can observe some correlation here, the mqtt message lenght is associated with the delay between packets, which makes sense since the DoS packets have higher message sizes and causes the server to send packets with higher delays, this is a good correlation to prove and feed it to a model
